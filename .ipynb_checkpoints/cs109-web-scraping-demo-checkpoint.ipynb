```python
#### Reading imdb movie csv into pandas

names = ['imdbid', 'title', 'year', 'score', 'vote', 'runtime', 'genres']
data = pd.read_csv('imdb_top_10000.txt', delimiter='\t', names=names).dropna()
print "Number of data rows %i" % data.shape[0]
data.head()
```

```python
## Turn runtime into a string

data['runtime'] = [float(r.split(' ')[0]) for r in data.runtime]

##To confirm change
data.head()
```

```python
data['title'] = [item[0:-7] for item in data.title]

data.head()
```

```python
genres = set()
for movie in data.genres:
    genres.update(movie.split('|'))
genres = sorted(genres)
```

```python
print genres
```

```python
#assign genres
for genre in genres:
    data[genre] = [genre in movie.split('|') for movie in data.genres]
data.head()
```

```python
## Most popular movies in descending order

print data[genres].sum().order(ascending=False)
```

```python
plt.figure(figsize=(20,3))
plt.scatter((data.score), data.vote, alpha=.3)
plt.yscale('log')
#plt.axis() for setting the axis ranges
plt.ylabel('Number of Votes')
plt.xlabel('Score out of 10')
plt.show()

```

```python
plt.figure(figsize=(20,3))
plt.scatter(data.vote,(data.score),alpha=.3)
plt.xscale('log')
#plt.axis() for setting the axis ranges
plt.xlabel('Number of Votes')
plt.ylabel('Score out of 10')
plt.show()
```

```python
##Highly voted on movies that had a low rating
##votes above 90.000 and less than rating 5

data[(data.vote > 90000) & (data.score <5)]
```

```python
data.groupby('year').score.mean()

##Plot mean score against the year
plt.figure(figsize=(20,3))
pd.Series.plot(data.groupby('year').score.mean(),alpha=.3)
#plt.axis() for setting the axis ranges
plt.xlabel('Year')
plt.ylabel('Mean Score of Movies out of 10')
plt.show()
```

```python

print "Number of movies from each year in each genre"
fig, axes = plt.subplots(nrows=4, ncols=6, figsize=(12,8), tight_layout=True) ##create all of the subplots
bins = np.arange(1950,2013,3) ## range of the years



for ax, genre in zip(axes.ravel(), genres): ##referring to every axes in each genre
    ax.hist(list(data[data[genre]==1].year), bins=bins, histtype='stepfilled', normed=True, color='r', alpha=.3, ec='none')
    ax.hist(data.year, bins=bins, histtype='stepfilled', ec='None', normed=True, zorder=0, color='0.75')
    ax.annotate(genre, xy=(1955, 3e-2), fontsize=14)
    ax.xaxis.set_ticks(np.arange(1950,2013,30))
    ax.set_yticks([])

    ax.set_xlabel('Year')
```
